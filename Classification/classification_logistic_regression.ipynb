```python
import pandas as pd
titanic_df = pd.read_csv('datasets/titanic_processed_3.csv')
titanic_df.head()
```

```python
titanic_df.shape
```

```python
from sklearn.model_selection import train_test_split

X = titanic_df.drop('Survived', axis=1)
Y = titanic_df['Survived']

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
x_train.describe()
```

```python
x_test.describe()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

# regularyzacja - ma zapobiec przetrenowaniu modelu (overfitting), uogólnia model. l1 - lasso regression, l2 - ridge regression
# polega na modyfikacji współczynników funkcji
# l1 sprawdza się jeśli w funkcji mamy wiele feature'ów - parametrów, może prowadzić do eliminacji niektórych parametrów
# l2 modyfikuje wszystkie współczynniki o tą samą wartość, nie doprowadza do eliminacji parametrów
penalty= 'l2' 

# parametr określający siłę regularyzacji - mniejsza wartość, to silniejsza regularyzacja
C = 1.0

# algorytm do optymalizacji, liblinear dobry dla małych zbiorów danych
solver = 'liblinear'

logistic_model = LogisticRegression(penalty = penalty, C = C, solver = solver).fit(x_train, y_train)
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
pred_results = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
```

```python
pred_results.head()
```

## Confusion matrix

```python
titanic_crosstab = pd.crosstab(pred_results.y_pred, pred_results.y_test)
titanic_crosstab
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

# (TP + TN) / (TP + FP + TN + FN)
print("accuracy: ", round(accuracy,2)) 

# TP / (TP + FP)
print("precision: ", round(precision,2)) 

 # TP / (TP + FN)
print("recall: ", round(recall,2))
```
